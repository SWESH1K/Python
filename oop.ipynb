```python
# % operator with negative numbers

print(5 % -3) # x = 5, y = -3
# --> r = x - (y * (x // y))
# --> r = 5  - ((-3) * (-2))
# --> r = 5 - (6) --> r = -1.
# It will print -1

print(-5 % 3) # x = -5, y = 3
# --> r = x - (y * (x // y))
# --> r = (-5)  - ((3) * (-2))
#  --> r = 5 - (-6) --> r = 1.
# It will print 1

print(-7 % 3)

print(7 % -3)


'''
To calculate the remainder r of dividing a number x by a number y, Python uses the equation r = x - (y * (x // y)).

For example, to find 5 % -3, Python first finds (5 // -3). Since 5 / -3 is about -1.67, that means 5 // -3 is -2.
 
Now Python multiplies that by -3 to get 6. Finally, Python subtracts 6 from 5 to get -1.

'''
```

```python
# Using List comprehension to find even numbers
import time

start=time.time()

limit = 1000000
even = [x for x in range(limit+1) if x%2==0]

end=time.time()

# print(even)
print("Time taken: ",format(end-start, ".4f"))
```

```python
#Not using List comprehension to find even numbers
import time

start=time.time()

limit = 1000000
even=[]
for i in range(limit+1):
    if(i%2==0):
        even.append(i)

# print(even)

end=time.time()

print("Time taken: ",format(end-start, ".4f"))
```

```python
# Using List comprehension to find even numbers and without set
import time

start=time.time()

limit = 1000000
num = [x for x in range(4,limit+1)]
even = [x for x in num if x%2==0]
odd = [x for x in num if x%2!=0]

end=time.time()

# print("Even:",even)
# print("Odd:",odd)
print("Time taken: ",format(end-start, ".4f"))
```

```python
# Using List comprehension to find even numbers and with set
import time

start=time.time()

limit = 1000000

num = [x for x in range(4,limit+1)]
even = [x for x in num if x%2==0]
odd = set(num) - set(even)

end=time.time()

# print("Even:",even)
# print("Odd:",odd)
print("Time taken: ",format(end-start, ".4f")) 
```

```python
'''
def <function_name>([parameters]);
    <statement(s)>
'''
```

def = The Keyword that informs Python that a function is being defined
<function_name> = A valid Python identifier that names the function
<parameters> = An optional, comma-separated list of parameters that may be passed to the function
<statements> = A set of valid Python lines

```python
def printN(a):
    if(a==0): return
    printN(a-1)
    return print(a,end=" ")
```

```python
def printReverseN(a):
    if(a==0): return
    print(a,end=" ")
    return printReverseN(a-1)
```

```python
# Concept of *args

def find_max(*args): # The symbol * is used to send infinte amount of parameters at the same time. 
    return max(args)

print(find_max(1,5,3,2)) # Output: 5
print(find_max(10,-5,2)) # Output: 10
```
